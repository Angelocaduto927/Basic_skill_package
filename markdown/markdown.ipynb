# Markdown Basic Language rules

---

## 标题格式

\# 一级标题

\## 二级标题

···

\###### 六级标题 （最多到6级）

---

## 横隔线   

\--- 在下方画一条横线   

---

## 字体样式 （下划线和*的作用是一样的）  

\**X\** （\_\_X\_\_） **加粗**   

\*X\* （\_X\_） *斜体*   

\~~X\~~ ~~删除线~~

\*\*\*X\*\*\* (\_\_\_X\_\_\_) ***粗斜体***

---

## 分段与换行

换行： 段落后面加两个空格

分段： 回车空一行

---

## 引用（可嵌套）

\> 引用
>引用1
>>引用2
>>>>>>>引用n

---

## 列表项

无序列表 \* （不与文本内容相连）

* 无序列表1
    * 子列表1 （tap+\*， 最多支持三级子列表）
* 无序列表2
* 无序列表3

有序列表 数字+. （以第一个数字为始，无论后续markdown列表数字是多少都是累加1; 如果要另开一个有序列表则需要要么分割行，要么另开一个markdown block）

1. 有序列表1
2. 有序列表2
100. 有序列表3

---

5. 有序列表1
6. 有序列表2
100. 有序列表3

---

## 表格

\|Column 1|Column 2|Column 3|  
\|-|-|-| （占位符有多少横杠都不影响，如果表格内容左对齐，在左边加":", |:---|,如果表格内容右对齐，在右边加":"，如果表格内容居中，在两边都加上）  
\|text 1|text 2|text 3| （中间有多少空格都不影响）

|Column 1|Column 2|Column 3|  
|-|-|-|
|张三|18|电工|
|李四|19|瓦工|
|小明|21|包工|
